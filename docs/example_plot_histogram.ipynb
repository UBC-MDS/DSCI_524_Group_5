# `EDAhelper.plot_histogram`

`plot_histogram` creates a repeated histogram plot for multiple numeric columns in a pandas DataFrame. To generate a plot using `plot_histogram`:

```python
# import function
from EDAhelper.plot_histogram import plot_histogram
import pandas as pd

# load data
df = pd.read_csv("https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/iris.data")
```

## Create histogram for all numeric columns in data

```python
plot_histogram(df)
```

## Create histogram for selected numeric columns in data

```python
plot_histogram(df, columns = ['SepalLength', 'SepalWidth'], num_bins = 10)
```

## Adjust histogram bin width for all plots

```python
plot_histogram(df, num_bins = 10)
```
